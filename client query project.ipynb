```python
# Import Required Libraries
```

```python
import sqlite3
import pandas as pd
from datetime import datetime
```

```python
# Create Database and Table
```

```python
conn = sqlite3.connect("query_db.db")
cursor = conn.cursor()


cursor.execute("""
CREATE TABLE IF NOT EXISTS queries (
query_id INTEGER PRIMARY KEY AUTOINCREMENT,
mail_id TEXT,
mobile_number TEXT,
query_heading TEXT,
query_description TEXT,
status TEXT,
query_created_time TEXT,
query_closed_time TEXT
)
""")


conn.commit()
print("Database and table created successfully")
```

```python
# Insert Client Query (Client Side Simulation)
```

```python
cursor.execute("""
INSERT INTO queries
(mail_id, mobile_number, query_heading, query_description, status, query_created_time)
VALUES (?, ?, ?, ?, ?, ?)
""", (
"test@gmail.com",
"9876543210",
"Login Issue",
"Unable to login to the application",
"Open",
datetime.now()
))


conn.commit()
print("Query inserted successfully")
```

```python
# View All Queries (Support Team View)
```

```python
df = pd.read_sql("SELECT * FROM queries", conn)
df
```

```python
# Close a Query (Support Team Action)
```

```python
cursor.execute("""
UPDATE queries
SET status = ?, query_closed_time = ?
WHERE query_id = ?
""", (
"Closed",
datetime.now(),
1
))


conn.commit()
print("Query closed successfully")
```
